```python
import pandas as pd 
```

```python
#a simple dummy data to learn merge() 
df1 = pd.DataFrame({'NAME':['White','Max','Evans','Richard','James','Kristos'],
                    'AGE':[12,23,61,19,27,11],
                    'EYE_COLOR':['Brown','Blue','Hazel','Green','Gray','Amber'], 
                    'HEIGHT(ft)':[1.45,2.34,6.7,4.5,2.45,8.1],
                    'COUNTRY': ['Singapore','Nigeria','Australia','US','Amsterdam','China']})

df2 = pd.DataFrame({'NAME':['Derick','Jack','Jane'],
                    'HIV/AIDS' :[1,0,1],
                    'BLOOD_TYPE' : ['A','AB','O'],
                    'COUNTRY': ['Singapore', 'Nigeria','China']})
```

```python
df1
```

```python
df2
```

```python
# merging on all columns 
df = pd.merge(df1,df2)
df
```

```python
#merging on specified columns 
df = pd.merge(df1,df2,on='COUNTRY')
df
```

```python
#left join 
df = pd.merge(df1,df2,on='COUNTRY', how='left')
df
```

```python
#right join 
df = pd.merge(df1,df2,on='COUNTRY',how='right')
df 
```

```python
#outer join 
df = pd.merge(df1,df2,on='COUNTRY',how='outer')
df
```
